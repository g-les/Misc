# Pattern Recognition for Malware Analysts & Threat Hunters

This reference page is intended to highlight a few common payload file headers when they are obfuscated with some typical encoding and encryption mechanisms. My hope is to show a comparison of them inline using identical data to help ingrain the patterns.

This is inspired by c3rb3ru5d3d53c's [Deobfuscating Scripts](https://c3rb3ru5d3d53c.github.io/documents/deobfuscating-scripts/) and Florian's [base64 cheat sheet](https://gist.github.com/Neo23x0/6af876ee72b51676c82a2db8d2cd3639)

If you can think of any other headers or encodings/encryptions one should train their eye to, submit a PR!
